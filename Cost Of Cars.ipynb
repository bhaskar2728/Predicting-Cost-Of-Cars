## Data Cleaning

### Removing the rows containing null values

### Splitting data into train and test set

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1,test_size = 0.2,random_state = 42)
split
```

```python
split.split(copy_data,copy_data['Year'])
```

```python
for train_index,test_index in split.split(copy_data,copy_data['Year']):
    strat_train_set = copy_data.loc[train_index]
    strat_test_set = copy_data.loc[test_index]
```

```python
strat_train_set = strat_train_set.dropna()
strat_test_set = strat_test_set.dropna()
```

```python
copy_data['Year'].value_counts()/len(copy_data)
```

```python
strat_test_set['Year'].value_counts()/len(strat_test_set)
```

```python
strat_train_set = strat_train_set.drop('Name',axis = 1)
strat_test_set = strat_test_set.drop('Name',axis = 1)
```

```python
train_data = strat_train_set.drop('Price',axis = 1)
train_labels = strat_train_set['Price']
test_data = strat_test_set.drop('Price',axis = 1)
test_labels = strat_test_set['Price']
```

### Using Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import OneHotEncoder,Imputer,StandardScaler

class DataFrameSelector(BaseEstimator,TransformerMixin):
    def __init__(self,attr_names):
        self.attr_names = attr_names
    def fit(self,X,y = None):
        return self
    def transform(self,X,y = None):
        return X[self.attr_names].values
    
num_attrs = ["Year","Kilometers_Driven","Seats","Mileage","Engine","Power"]
text_attrs = ["Location","Fuel_Type","Transmission","Owner_Type"]


num_pipeline = Pipeline([('selector',DataFrameSelector(num_attrs)),('imputer',Imputer(strategy = "median")),('std_scalar',StandardScaler()),])
text_pipeline = Pipeline([('selector',DataFrameSelector(text_attrs)),('one_hot_encoder',OneHotEncoder(sparse = False)),])
```

```python
from sklearn.pipeline import FeatureUnion

full_pipeline = FeatureUnion(transformer_list = [('num_pipeline',num_pipeline),('text_pipeline',text_pipeline),])
```

```python
train_data_prepared = full_pipeline.fit_transform(train_data)
```

## Training our model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(train_data_prepared,train_labels)
```

```python
some_data = train_data.iloc[:5]
some_labels = train_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print(lin_reg.predict(some_data_prepared))
print(list(some_labels))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
dataset_predictions = lin_reg.predict(train_data_prepared)
lin_mse = mean_squared_error(train_labels,dataset_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(train_data_prepared,train_labels)
```

```python
dataset_predictions = tree_reg.predict(train_data_prepared)
tree_mse = mean_squared_error(train_labels,dataset_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(train_data_prepared,train_labels)
dataset_predictions = forest_reg.predict(train_data_prepared)
forest_mse = mean_squared_error(train_labels,dataset_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(forest_reg,train_data_prepared,train_labels,scoring ='neg_mean_squared_error' ,cv = 10)
forest_reg_rmse_scores = np.sqrt(-scores)
```
